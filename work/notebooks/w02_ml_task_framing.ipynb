## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking. I choose ranking because my goal is to prioritize a list of existing pages by their potential to attract AI-referred traffic, rather than simply labeling them as "good" or "bad".

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

My target is the ratio of AI-referred sessions to total sessions over the last 90 days. This is an observed outcome from the data, serving as a proxy for how well a page is currently positioned to be surfaced by AI models.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Precision@K (e.g., Precision@20). This metric is defensible because it directly measures if the top-ranked pages we suggest for optimization are indeed those that show high potential for AI growth. A "good" number depends on the client's baseline, but significantly outperforming a random selection is the target.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

The pattern is too messy for a fixed rule because AI-referred traffic is influenced by a complex interplay of content structure, domain authority, and freshness. An "if-statement" (e.g., "if sessions > 100") would miss pages with high thematic relevance but lower volume, whereas a ranking model can learn these nuanced signals across thousands of pages.
